```python
# Trabalhando com a biblioteca Pandas
#
#
#

import pandas as pd

aluguelCasa = pd.read_csv("/content/House_Rent_Dataset.csv")  # Importando o dataset em csv
```

```python
# Renomeando colunas com .rename()

aluguelCasa.rename(columns={'Area Type':'area_type', 'Area Locality':'area_locality', 'Furnishing Status':'furnishing_status', 'Tenant Preferred':'tenant_preferred', 'Point of Contact':'point_of_contact'})
```

```python
# Vendo o tipo de dado de cada coluna com dtypes

aluguelCasa.dtypes
```

```python
# Alterando o tipo de dado de uma coluna com astype()

aluguelCasa = aluguelCasa.astype({'Rent':'float'})

aluguelCasa.dtypes
```

```python
# Usando .info

aluguelCasa.info()
```

```python
# Pegando apenas as colunas númericas

aluguelCasa[['BHK','Rent','Size','Bathroom']]
```

```python
# Análise descritiva com .describe()

aluguelCasa[['BHK','Rent','Size','Bathroom']].describe()
```

```python
# Filtrando dados
# Aplicado filtro para aluguel menor OU igual a 4000

aluguelCasa_filtro = aluguelCasa[(aluguelCasa['Rent'] <= 4000)]

aluguelCasa_filtro.head()
```

```python
# Aplicando dois filtros
# Aplicando filtro para o aluguel e mais do que 1 banheiro

aluguelCasa_filtro = aluguelCasa[(aluguelCasa['Rent'] <= 4000) & (aluguelCasa['Bathroom'] > 1)]

aluguelCasa_filtro.head()
```

```python
# Aplicando filtro de OU |

aluguelCasa_filtro = aluguelCasa[(aluguelCasa['Rent'] == 4000) | (aluguelCasa['Size'] >= 200) & (aluguelCasa['Bathroom'] > 1)]

aluguelCasa_filtro.head()
```

```python
# Puxando um novo describe

aluguelCasa.describe()
```

```python
aluguelCasa = aluguelCasa.query('Rent <= 7000 and Bathroom >1')
aluguelCasa
```

```python
# Agregando com agg e groupby
agg_city_price = aluguelCasa.groupby(['City'])['Rent'].agg('mean')
agg_city_price

# Utilizando o reset_index

agg_city_price = aluguelCasa.groupby(['City'])['Rent'].agg('mean').reset_index()
agg_city_price
```

```python
# Groupby bathrooms

agg_bathroom = aluguelCasa.groupby(['Bathroom'])['Rent'].agg('mean').reset_index()
agg_bathroom
```

```python
# Importando Seaborn

import seaborn as sns
```

```python
# Plotando gráfico de barras

ax = sns.barplot(data = agg_bathroom, x = 'Bathroom', y = 'Rent')

ax.set(xlabel = 'Qtd. Banheiros', ylabel = 'Valor Aluguel');
```

```python
# Utilizando hue no histplot

aluguelCasa = aluguelCasa.query('Size <= 2500')

hx2 = sns.histplot(data = aluguelCasa, x = 'Size', hue = 'Bathroom')
```

```python
# Plotando histograma

aluguelCasa = aluguelCasa.query('Size > 0 and Size < 2001')

agg_size = aluguelCasa[['Size']]

hx = sns.histplot(data = agg_size, x = 'Size')
```

```python
# Plotando histograma com rent

aluguelCasa = aluguelCasa.query('rent')

agg_size = aluguelCasa[['Size']]

hx = sns.histplot(data = agg_size, x = 'Size')
```

```python
# Puxando um .describe() do Rent para visualizar a distribição dos valores de aluguel

aluguelCasa['Rent'].describe()
```

```python
# Calculando o limite superior a partir do intervalo interquartil e aplicando um teto no valor da renda para eliminar outliers:

iqr = 19000 # Q3 - Q1 sendo Q3 = 29000 e Q1 = 10000

limiteRent = 29000 + (1.5 * iqr)
```

```python
aluguelCasa = aluguelCasa.query(f"Rent <= {limiteRent}")

agg_rent = aluguelCasa[['Rent']]

hx2 = sns.histplot(data = aluguelCasa, x = 'Rent', hue = 'Bathroom', binwidth = 2000, palette = 'viridis')

hx2.set(xlabel = 'Aluguel', ylabel = 'Quantidade');
```
